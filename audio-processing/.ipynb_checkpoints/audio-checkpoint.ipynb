```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.io import wavfile
from scipy.signal import convolve
from sklearn.preprocessing import normalize
from IPython.display import Audio as display_audio

%matplotlib inline
```

```python
fs_pist, pistol = wavfile.read('data/pistol.wav')
fs_viol, violin = wavfile.read('data/violin.wav')
```

Let's treat the gun shot recording as an *impulse response* of the acoustic environment, viewed as an LTI system $ S $. If $ S $ really is LTI, then the system is completely characterized by its impulse response. If we convolve our violin signal with the gun shot, the output is in effect what the violin *would have sounded like* had it been recorded in the same room as the gun shot.

```python
violin_pistol = convolve(violin, normalize(pistol, axis=0), 'same')
```

```python
def get_time(signal, fs):
    N = signal.shape[0]
    t = np.linspace(0,  N/fs, N)
    return t
```

```python
f, axarr = plt.subplots(3, 1)

t = get_time(pistol, fs_pist)
axarr[0].plot(t, pistol[:, 0])
axarr[0].set_title('Pistol')

t = get_time(violin, fs_viol)
axarr[1].plot(t, violin[:, 0])
axarr[1].set_title('Violin')

t = get_time(violin_pistol, fs_viol)
axarr[2].plot(t, violin_pistol[:, 0])
axarr[2].set_title('Violin convolved with pistol')

for ax in axarr:
    ax.get_yaxis().set_visible(False)
    ax.set_xlabel('Time (s)')
    
plt.tight_layout()
plt.show()
```

```python
display_audio(data=np.transpose(pistol), rate=fs_pist)
```

```python
display_audio(data=np.transpose(violin), rate=fs_viol)
```

```python
display_audio(data=np.transpose(violin_pistol), rate=fs_viol)
```
