```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import ndimage
from scipy.signal import convolve

%matplotlib inline
```

```python
def plot_pandas(panda1, panda2):
    fig, axarr = plt.subplots(nrows=1, ncols=2, sharey=True)
    axarr[0].imshow(panda1)
    axarr[1].imshow(panda2)
    for ax, title in zip(axarr, ['panda', 'panda_lpf']):
        ax.set_title(title)
    plt.suptitle('Pandas')
    plt.show()
```

```python
panda = ndimage.imread('panda.jpg')[:, :, 0]

cx, cy = (c // 2 for c in panda.shape) 
delta = 1

row_idc = slice(cx - delta, cx + delta)
col_idc = slice(cy - delta, cy + delta)

lpf = np.zeros_like(panda)
lpf[row_idc, col_idc] = 1

panda_lpf = convolve(panda, lpf, 'same')
plot_pandas(panda, panda_lpf)
```
